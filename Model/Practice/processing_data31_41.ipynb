```python
import numpy as np
import pandas as pd
```

## data31 data41是原始数据（data31_target, phone, target）

```python
df1 = pd.read_csv('datasets/data31.csv')
df2 = pd.read_csv('datasets/data41.csv')
```

## DataFrame去重（去掉phone列相同的行，保留第一行）

```python
# df1 = pd.read_csv('datasets/data31.csv')
# df2 = pd.read_csv('datasets/data41.csv')

norepeat_df1 = df1.drop_duplicates(subset='phone', keep='first')
norepeat_df2 = df2.drop_duplicates(subset='phone', keep='first')

len(norepeat_df1), len(norepeat_df2)
```

## phone -> id

```python
phone1 = list(norepeat_df1["phone"])
p1 = {}
for i in range(len(phone1)):
    p1[phone1[i]] = i
    
phone2 = list(norepeat_df2["phone"])
p2 = {}
for i in range(len(phone1), len(phone1)+len(phone2)):
    p2[phone2[i-len(phone1)]] = i
```

```python
p1
```

```python
p2
```

## phone相同时id变为相同

```python
for key, _ in p2.items():
    if key in p1:
        p2[key] = p1[key]
p2
```

## 取出phone对应的id

```python
id1 = list(p1.values())
```

```python
id2 = list(p2.values())
```

```python
len(id1)
```
