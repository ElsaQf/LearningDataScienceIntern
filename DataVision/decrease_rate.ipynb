```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
%matplotlib inline
```

```python
date_count = pd.read_csv('test_3.csv')
```

```python
date_count.head()
```

```python
fig = plt.figure(figsize=(6,3))
plt.plot(range(1,32), date_count.Count_Id)
```

```python
plt.bar(range(1,32), date_count.Count_Id, alpha=0.5, width=0.5, color='blue', label='The First Bar', lw=3)
```

```python
count_31 = date_count['Count_Id']
```

```python
decrease = []
for i in range(30):
    j = count_31[i+1] - count_31[i]
    decrease.append(j)
```

```python
decrease
```

```python
fig = plt.figure(figsize=(6,3))
plt.plot(range(30), decrease)
```

```python
decrease_rate = []
for i in range(30):
    j = (count_31[i+1] - count_31[i]) / count_31[i]
    decrease_rate.append(j)
fig = plt.figure(figsize=(6,3))
plt.plot(range(30), decrease_rate)
```
