# 爬虫 Spider with urllib

```python
from urllib import request
```

## urllib.request
## 以http://fanyi.baidu.com/ 爬取

### urllib.request
### response = request.urlopen()
### html = response.read()

### html = html.decode("utf-8")

```python
response = request.urlopen("http://fanyi.baidu.com/")
html = response.read()
print(html)
```

```python
html = html.decode('utf-8')
print(html)
```

```python
import chardet

response = request.urlopen("http://fanyi.baidu.com/")
html = response.read()
charset = chardet.detect(html)
print(charset)
```
