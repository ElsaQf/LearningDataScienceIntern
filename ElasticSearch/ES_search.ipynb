## ES查找，导出csv

```python
import pandas as pd
from elasticsearch import Elasticsearch

es = Elasticsearch('http://dev.rrx.es01.srv:9200/')
result = es.search(index='user_info', size=100)

source_user_info = pd.DataFrame(pd.DataFrame(result['hits']['hits'])['_source'].to_dict()).T
```

```python
source_user_info
```

```python
file = open('file_user_info.csv','w',encoding='utf-8')
source_user_info.to_csv(file, index=None)
file.close()
```

```python
result = es.search(index='jdb_overdue_data', size=100)
source_jdb_overdue_data = pd.DataFrame(pd.DataFrame(result['hits']['hits'])['_source'].to_dict()).T
file = open('file_jdb_overdue_data.csv', 'w', encoding='utf-8')
source_jdb_overdue_data.to_csv(file, index=None)
file.close()
```
